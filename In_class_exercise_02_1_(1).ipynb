Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://www.imdb.com/search/title/?title_type=feature' #Input the url
reqs = requests.get(url)
url_data = BeautifulSoup(reqs.text, 'lxml') #To get the data the URL
url_tags = url_data.find_all('a', attrs={'class': None})
url_tags_data = [tag.attrs['href'] for tag in url_tags 
              if tag.attrs['href'].startswith('/title') & tag.attrs['href'].endswith('/')]
url_tags_data = list(dict.fromkeys(url_tags_data)) 
url_link = ["https://www.imdb.com" + tag + 'reviews' for tag in url_tags_data] # Combine the tags n Url to get the link
print(url_link)
print(url_link[0])
print(url_link[1])

def getindex(a):
    minpos = a.index(min(a))
    maxpos = a.index(max(a))
    return minpos,maxpos

comment = []
vote = []
vote1=[]
UID=[]
cinema=[]
review_note=[]
for i in range(0,50):
    url_data = BeautifulSoup(requests.get(url_link[i]).text, 'html.parser')
    voting_per = [tag.previous_element for tag in 
                           url_data.find_all('span', attrs={'class': 'point-scale'})] #Rating percentage
    if len(voting_per) > 0:   #If rating percentage is greater than 0
        n_index, p_index = getindex(list(map(int, voting_per)))
        review_list = url_data.find_all('a', attrs={'class':'title'})
        reviews = ["https://www.imdb.com" + review['href'] for review in review_list]
    else:
        print(None)



    for j in reviews:
        text=BeautifulSoup(requests.get(j).text, 'html.parser').find('div', attrs={'class': 'text show-more__control'}).get_text()
        rat=BeautifulSoup(requests.get(j).text, 'html.parser').find('div', attrs={'class': 'lister-item-content'}).get_text()
        movie_user=(BeautifulSoup(requests.get(j).text, 'html.parser').find('title').get_text())
        short_des=(BeautifulSoup(requests.get(j).text, 'html.parser').find('a', attrs={'class': 'title'}).get_text())
        len_rat1=len(vote1)
        if(text == ''):
            comment.append('No Review')
        else:
            comment.append(text)
        vote.append(rat)

        UID.append(movie_user.split("'s")[0])
        cinema.append(movie_user.split('Review of')[1])
        review_note.append(short_des)
  
    print('vote : '+str(len(vote))+' review list:'+str(len(comment)))
    if(len_rat1==0):
        k=0
    else:
        k=len_rat1
    for j in range(k,len(vote)):
        vote1.append(vote[j].split('/10')[0].split('\n')[len(vote[j].split('/10')[0].split('\n'))-1])
    print('rating1:'+str(len(vote1)))

data = {
        "Cinema name" : cinema,
        "User Name" : UID,
        "Review_Note" : review_note,
        'Detail Reviews': comment,
        'Rating': vote1,
        
        }
df = pd.DataFrame(data)
df

```
